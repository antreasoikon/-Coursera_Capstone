Load the wikipedia page in html format

```python
response = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
```

Extract the table 

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
all_tables=soup.find_all("table")
all_tables
```

```python
right_table=soup.find('table', class_='wikitable sortable')
right_table
```

```python
A=[]
B=[]
C=[]

for row in right_table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==3:
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
```

```python
import pandas as pd
df=pd.DataFrame(A,columns=['Postcode'])
df['Borough']=B
df['Neighbourhood']=C
df.head()
```

Clean the data

```python
df = df[~(df['Borough']== 'Not assigned')]
```

```python
df.Neighbourhood.replace("Not assigned", df.Borough, inplace=True)
```

```python
df_g = df.groupby(['Postcode','Borough'], sort = False)['Neighbourhood'].apply(lambda x: ', '.join(x)).apply(lambda x: x.replace('\n', '')).reset_index()
```

Get the geolocation

```python
radius=500
LIMIT=100
```

```python
CLIENT_ID = 'YYXPDRNWZJTT1G2FQJUB411VM1ICD33JHQ0QSXBELFS41AOW' 
CLIENT_SECRET = 'XLAJZA5WJ3CPFBJGJNEBY02N5ULTLCGM5VI0KMOT03NUJQZ5' 
VERSION = '20200316'
```

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighbourhood', 
                  'Neighbourhood Latitude', 
                  'Neighbourhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
toronto_venues = getNearbyVenues(names=df_tor['Neighbourhood'],
                                   latitudes=df_tor['Latitude'],
                                   longitudes=df_tor['Longitude']
                                  )
```

```python
toronto_venues.head()
```

One hot encode  Venue categories 

```python
toronto_onehot = pd.get_dummies(toronto_venues['Venue Category'],prefix="", prefix_sep=" ")
toronto_onehot['Neighbourhood'] = toronto_venues['Neighbourhood']
```

```python
toronto_onehot.head()
```

```python
toronto_onehot.shape
```

Take the mean of the one hot vectors per Neighbourhood

```python
onehot_group = toronto_onehot.groupby('Neighbourhood').mean().reset_index()
```

```python
onehot_group.head()
```

Cluster Neighbours using KMeans algorithm

```python
from sklearn.cluster import KMeans
```

```python
# set number of clusters
kclusters = 5

toronto_grouped_clustering = onehot_group.drop('Neighbourhood', axis=1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)

kmeans.labels_[0:10]
```

```python
onehot_group['labels'] = kmeans.labels_
```

```python
onehot_group.head()
```

Create the final merged dataframe

```python
df_fin = pd.merge(toronto_venues, onehot_group[['Neighbourhood','labels']], on ='Neighbourhood')
```

```python
df_fin.head()
```

```python
address = 'Toronto, CA'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Manhattan are {}, {}.'.format(latitude, longitude))
```

```python
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)


# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(df_fin['Venue Latitude'], df_fin['Venue Longitude'],df_fin['Neighbourhood'], df_fin['labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
